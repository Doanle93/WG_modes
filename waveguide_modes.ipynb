```python
import matplotlib.pyplot as plt
import numpy as np
import gplugins.tidy3d as gt
```

```python
## FINDING MODES OF A STRIP WAVEGUIDE
strip = gt.modes.Waveguide(
    wavelength = 1.55,
    core_material = "si",
    core_thickness = 0.22,
    core_width = 0.44,
    clad_material = "sio2",
)
strip.plot_index()
```

```python
# show TE first mode
strip.plot_field(field_name= "Ex", mode_index = 0) 
```

```python
# show TM mode
strip.plot_field(field_name= "Ey", mode_index= 1)
```

```python
## FINDING MODES OF A RIB WAVEGUIDE
rib = gt.modes.Waveguide(
    wavelength= 1.55,
    core_material="si",
    core_thickness= 0.22,
    core_width= 0.5,
    slab_thickness= 0.09,
    clad_material= "sio2", 
)
rib.plot_index()
```

```python
# show effective refractive index
rib.n_eff
```

```python
# show TE mode
rib.plot_field(field_name= "Ex", mode_index= 0)
```

```python
## FINDING MODES OF A RIB WAVEGUIDE WITH A SUBSTRATE

rib_s = gt.modes.Waveguide(
    wavelength= 1.55,
    core_material="si",
    core_thickness= 0.22,
    core_width= 0.5,
    slab_thickness= 0.09,
    clad_material= 1, 
    clad_thickness= 0.8,
    box_material= "sio2",
    box_thickness= 0.5,
)
rib_s.plot_index()
```

```python
# show TE mode
rib.plot_field(field_name= "Ex", mode_index= 0)
```

```python
# print effective refractive index
rib_s.n_eff
```
