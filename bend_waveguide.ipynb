```python
import matplotlib.pyplot as plt
import numpy as np
import gplugins.tidy3d as gt
```

```python
radii = np.arange(4, 7)
bend = gt.modes.Waveguide(
    wavelength = 1.55,
    core_material= "si", 
    core_width= 0.5,
    core_thickness= 0.22,
    clad_material="sio2",
    clad_thickness= 1.5,
    box_material= "sio2",
    box_thickness=1.5,
    num_modes= 1,
    bend_radius= radii.min(),
)

bend.plot_index()
```

```python
mismatch = gt.modes.sweep_bend_mismatch(bend, radii)

plt.plot(radii, 10 * np.log10(mismatch))
plt.xlabel("Radius (m)")
plt.ylabel("Mismatch (dB)")
```
