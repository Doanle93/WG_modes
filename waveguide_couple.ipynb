```python
import matplotlib.pyplot as plt
import numpy as np

import gplugins.tidy3d as gt
```

```python
c = gt.modes.WaveguideCoupler(
    wavelength=1.55,
    core_width=(0.5, 0.5),
    gap=0.1,
    core_thickness=0.22,
    slab_thickness=0.09,
    core_material="si",
    clad_material="sio2",
)
c.plot_index()
```

```python
c.plot_field(field_name= "Ex", mode_index=1)
```

```python
coupler = gt.modes.WaveguideCoupler(
    wavelength=1.55,
    core_width=(0.5, 0.5),
    core_thickness=0.220,
    slab_thickness=0.09,
    core_material="si",
    clad_material="sio2",
    num_modes=4,
    gap=0.1,
)

print("\nCoupler:", coupler)
print("Effective indices:", coupler.n_eff)
print("Mode areas:", coupler.mode_area)
print("Coupling length:", coupler.coupling_length())

gaps = np.linspace(0.05, 0.4, 21)
lengths = gt.modes.sweep_coupling_length(coupler, gaps)

_, ax = plt.subplots(1, 1)
ax.plot(gaps, lengths)
ax.set(xlabel="Gap (μm)", ylabel="Coupling length (μm)")
ax.legend(["TE", "TM"])
ax.grid()
```
