```python
import matplotlib.pyplot as plt
import numpy as np
import gplugins.tidy3d as gt
```

```python
## WAVEGUIDE WIDTH EFFECTIVE INDEX @ 1550 nm
strip = gt.modes.Waveguide(
    wavelength = 1.31,
    core_material = "si",
    core_thickness = 0.22,
    core_width = 1,
    slab_thickness= 0.0,
    clad_material = "sio2",
    num_modes= 4,
)
strip.plot_index()
```

```python
w = np.linspace(0.1, 0.8, 7)
n_eff = gt.modes.sweep_n_eff(strip, core_width = w)
fraction_te = gt.modes.sweep_fraction_te(strip, core_width = w)

for i in range(4):
    plt.plot(w, n_eff.sel(mode_index = i).real, color = "blue")
    plt.scatter(w, n_eff.sel(mode_index = i).real, color = 'k')

plt.axhline(y = 1.44, color = 'k', ls = '--')
plt.colorbar().set_label("TE fraction")
plt.xlabel("Width of waveguide (µm)")
plt.ylabel("Effective refractive index")
plt.title("Effective index sweep")
```
